```python
%%cython
import cython

cimport numpy as cnp


'''
implement the dot product function:
This function calculates the scalar product of two vector
'''

@cython.boundscheck(False)
@cython.wraparound(False)
def dot_product(cnp.ndarray[double, ndim=1] A, cnp.ndarray[double, ndim=1] B):
    cdef int i
    cdef double result = 0
    for i in range(A.shape[0]):
        result += A[i] * B[i]
    return result
```

вычисление скалярного произведения двух векторов, заданных как два массива чисел

случайный вектор генерируется через новый метод random.default_rng(), а не через старый вариант - который скоро уберут - random.rand()

```python
import numpy as np

rng = np.random.default_rng()
A = rng.random(100000000)
B = rng.random(100000000)
%timeit dot_product(A, B)
```
