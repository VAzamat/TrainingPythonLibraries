```python
%%cython
import cython
from libc.stdint cimport int64_t


@cython.boundscheck(False)
@cython.wraparound(False)
def factorial(int n):
    cdef int i
    cdef int64_t result = 1
    for i in range(1, n + 1):
        result *= i
    return result
```

для заданной реализации функции возможно вычисление до 20!. При вычислении 21! возникает переполнение 64-bit значение и надо использовать тип double - тогда появиться возможность считать до 170! (будет понятен порядок числа, но при этом теряется точность последних знаков) 

```python
import numpy as np
print( factorial(18) )
```
