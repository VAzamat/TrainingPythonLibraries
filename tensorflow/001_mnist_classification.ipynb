```python
import tensorflow as tf
from tensorflow.keras import layers, models, datasets
```

```python
(x_train, y_train), (x_test, y_test) = datasets.mnist.load_data()
x_train, x_test = x_train / 255.0, x_test / 255.0 
```

```python
#create the model
model = models.Sequential([
    layers.Flatten(input_shape=(28, 28)), # 袩褉械芯斜褉邪蟹芯胁邪薪懈械 2D 胁械泻褌芯褉邪 胁 1D
    layers.Dense(128, activation='relu'), # 袩芯谢薪芯褋胁褟蟹薪褘泄 褋谢芯泄 褋 128 薪械泄褉芯薪邪屑懈
    layers.Dropout(0.2), # Dropout 写谢褟 褉械谐褍谢褟褉懈蟹邪褑懈懈
    layers.Dense(10, activation='softmax') # 袙褘褏芯写薪芯泄 褋谢芯泄 (10 泻谢邪褋褋芯胁)
])
```

```python
#compile model
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
#training the model
model.fit(x_train, y_train, epochs=5, batch_size=32, validation_split=0.1)
```

```python
#estimate the accuracity
test_loss, test_acc = model.evaluate(x_test, y_test)
print(f"孝芯褔薪芯褋褌褜 薪邪 褌械褋褌芯胁褘褏 写邪薪薪褘褏: {test_acc}")
```
