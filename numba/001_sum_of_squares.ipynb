```python
from numba import jit
import numpy as np
```

описываем функцию и используем декоратор, чтобы заставить скомпилировать в быстрый код

```python
@jit(nopython=True)
def sum_of_squares(arr):
    result = 0.0
    for x in arr:
        result += x * x
    return result
```

определяем массив

```python
arr = np.random.randn(10000000)
```

запускаем первый раз, чтобы время компиляции не учитывалось при подсчете времени выполнения


```python
_ = sum_of_squares(arr[:10])
```

теперь делаем точный замер скорости выполнения

```python
%timeit sum_of_squares(arr)
```
