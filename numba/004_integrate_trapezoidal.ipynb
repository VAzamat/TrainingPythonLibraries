```python
from numba import jit
import numpy as np
```

```python
@jit(nopython=True)
def f(x):
    # f(x) = x^2 + 2x + 1
    return x * x + 2.0 * x + 1.0


@jit(nopython=True)
def integrate_trapezoidal_numba(a, b, n):
    """
    Вычисляет определенный интеграл функции f(x) на отрезке [a, b] 
    методом трапеций с разбиением на n отрезков.
    """
    if n <= 0:
        raise ValueError("Количество разбиений n должно быть больше нуля.")
        
    h = (b - a) / n 
    total_sum = 0.5 * (f(a) + f(b))
    
    for i in range(1, n):
        x = a + i * h
        total_sum += f(x)
        
    return total_sum * h
```

```python
%timeit integrate_trapezoidal_numba(0.0, 10.0, 1000000)
```
