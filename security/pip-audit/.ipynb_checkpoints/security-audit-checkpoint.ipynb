pip-audit проверяет по открытой базе данных Python Advisory Database какие из приложений уязвимы.

запуск для проверки библиотек из локального окружения:

```python
!pip-audit
```

вывод программы **`nltk 3.9.4   PYSEC-2026-597`** символизирует наличие уязвимости

Комментарий:
PYSEC-2026-597 — это идентификатор уязвимости Path Traversal (обход дерева каталогов) в библиотеке NLTK версии 3.9.4. Проблема кроется в механизме загрузки ресурсов через функцию nltk.data.load(). Причина: Встроенная проверка безопасности (регулярное выражение _UNSAFE_NO_PROTOCOL_RE) проверяет путь на наличие опасных символов вроде ../ до того, как декодируется URL-последовательность.
